```python
dummy_csv = pd.read_csv('./dummy - 10 lines.csv')

dummy_csv
```

```python
dummy_csv = dummy_csv.fillna('Nothing')
```

```python
print(dummy_csv['argument 1'][0])
print(dummy_csv['argument 1'][2])
```

```python
sequence_dict = {
    'sequence' : [],
    'api start' : [],
    'api list' : [],
    'which argument' :[],
    'which argument index' : []
}
total_argument = 4
```

```python
index = 0
seq = 1
k=1
for api in dummy_csv['api']:
#     index = 0
    api_list = []
    arg_list = []
    arg_index_list = []
#     api_list.append(api)
    print(api_list)
    print(api)
    for args in range(total_argument):
        
        arg = dummy_csv['argument ' + str(args+1)][index]
        print('=============== LAGI DI ', arg, '===========')
        if arg == 'Nothing':
            print('SORRY ini nothing, ga bisa')
#             index += 1
            continue
        for i in range(len(dummy_csv['argument 1'])):
    #         print(i+1)
            if i < index:
                print('index', i+1, 'ga bisa ya')
                continue
    #         print(arg , dummy_csv['argument 1'][i])
            for j in range(total_argument):
                print('argument ' + str(j+1))
                print(arg, 'bandingkan dengan', dummy_csv['argument ' + str(j+1)][i])
                if dummy_csv['argument ' + str(j+1)][i]=='Nothing':
    #                 print('NOTHING nih ngab')
                    continue
        
                if arg==dummy_csv['argument ' + str(j+1)][i]:
                    for k in range(total_argument):
                        arg2 = dummy_csv['argument ' + str(k+1)][i]
#                         print('arg2 --> ',arg2)
                        if arg2 == 'Nothing':
                            continue

                        if arg2 == dummy_csv['argument ' + str(j+1)][i]:
                            
                            api_list.append(dummy_csv['api'][i])
                            arg_list.append('argument ' + str(j+1))
                            arg_index_list.append(i)
                            print('YEY sama guys')
            #                 print(api_list)
                        else:
                            print('WADUH gak sama guys')
                        
            if i+1 ==len(dummy_csv['argument 1']):
                print('mentok bos')
                print('sequence :',seq)
                print('api_start :', api)
                print('api :', api_list)
                sequence_dict['sequence'].append(('sequence ' + str(seq)))
                sequence_dict['api start'].append(api)
                sequence_dict['api list'].append(api_list)
                sequence_dict['which argument'].append(arg_list)
                sequence_dict['which argument index'].append(arg_index_list)
                api_list = []
                arg_list = []
                arg_index_list = []
#                 api_list.append(api)
    
    seq+=1
    index += 1
```
